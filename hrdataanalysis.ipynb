# AGGREGATION AND GROUPING

### PANDAS

```python
df['Salary'].sum()
```

```python
df['Absences'].mean()
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Salary'].agg(['mean', 'sum', 'max', 'min', 'std'])
```

```python
df.groupby('PerformanceScore').agg(
    Highest_salary=('Salary','max'),
    Average_Salary=('Salary','mean')
)
```

```python
df.groupby(['Department','Sex']).agg({'Salary':['mean','max','min'],'Absences':['sum']})
```

# TRANSFORMATIONS

### PANDAS

```python
df['Salary_in_K']=df['Salary']/1000
```

```python
today = pd.Timestamp.today()
df['YearsWorkedTogether'] = df['DateofHire'].apply(lambda doh: today.year - doh.year - ((today.month, today.day) < (doh.month, doh.day)))
```

### NUMPY

```python
df['EngagementSurvey_Category'] = np.where(df['EngagementSurvey']>3, 'Good', 'Bad')
```

```python
np.unique(df['EngagementSurvey_Category'])
```

```python
df['EngagementSurvey_Category'].value_counts()
```
